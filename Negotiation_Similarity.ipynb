```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_md")

# Sample sentences
sentence1 = "transaction value will be 0.4"
sentence2 = "Ok , Done, Lets now discuss about the delivery time"

# Process the sentences with spaCy
doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

# Calculate semantic similarity
semantic_similarity = doc1.similarity(doc2)

print(f"Semantic Similarity between Sentence 1 and Sentence 2: {semantic_similarity}")
```

```python
data = {
    "Cand" : X['data'],
    "Emp": Y['data'],
    "output": X['output'] + Y['output'],
    "offer" : X['Offers'] + Y['Offers'],
    "Accept":X['Accept'] + Y['Accept'],
    "Query":X['Query'] + Y['Query'],
    "Reject":X['Reject'] + Y['Reject'],
    "Greet":X['Greet'] + Y['Greet'],
}
```

```python
df1 = pd.DataFrame(data)
```

```python
length = len(X)-len(Y)
length
```

```python
len_df = len(df1)
len_df
```

```python
lsts = []
for i in range(len_df-1,len_df-length-1,-1):
  print(i)
  lsts.append(i)
```

```python
df1.drop(lsts,axis=0,inplace=True)
```

```python
df1
```

```python
lsts_sim = []

for i in range(0,len(df1)):
  # Process the sentences with spaCy
  doc1 = nlp(df1['Cand'][i])
  doc2 = nlp(df1['Emp'][i])

  # Calculate semantic similarity
  semantic_similarity = doc1.similarity(doc2)
  lsts_sim.append(semantic_similarity)
```

```python
lsts_sim
```

```python
df1['Similarity_Score'] = lsts_sim
```
